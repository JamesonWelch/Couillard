```python
from ast import literal_eval
# literal_eval('[1.23, 2.34]')

def extract_phone_numbers(string):

    r = re.compile(r'(\d{3}[-\.\s]??\d{3}[-\.\s]??\d{4}|\(\d{3}\)\s*\d{3}[-\.\s]??\d{4}|\d{3}[-\.\s]??\d{4})')
    phone_numbers = r.findall(string)
    numbers =  [re.sub(r'\D', '', number) for number in phone_numbers]
    extracted_numbers = []
    [extracted_numbers.append(number) for number in numbers if number not in extracted_numbers]
    return extracted_numbers

def de_list(l):
    if len(l) > 1:
        literal_eval(l)
        
```

```python
def extract_email_addresses(string):

    r = re.compile(r'[\w\.-]+@[\w\.-]+')
    emails = []
    [emails.append(item) for item in r.findall(string) if item not in emails]
    return emails
```

```python
import string

def ie_preprocess(document):

    
    document = ' '.join([i for i in document.split() if i not in stop])
    sentences = nltk.sent_tokenize(document)
    sentences = [nltk.word_tokenize(sent) for sent in sentences]
    sentences = [nltk.pos_tag(sent) for sent in sentences]
    return sentences


def extract_names(document, address):


    names = []
    sentences = ie_preprocess(document)
    for tagged_sentence in sentences:
        for chunk in nltk.ne_chunk(tagged_sentence):
            if type(chunk) == nltk.tree.Tree:
                if chunk.label() == 'PERSON':
                    names.append(' '.join([c[0] for c in chunk]))
    extracted_names = []
    [extracted_names.append(name) for name in names if name not in extracted_names]

#     names_checked = []

#     for name in name_list:
#         for item in extracted_names:
#             if name in item.lower():
#                 if item not in names_checked:
#                     names_checked.append(item)

#     first_names = [x.split()[0] for x in names_checked]
# #     print(f'NAMES_CHECKED BEFORE CLEANING: {names_checked}')

#     for name in names_checked:
#         if name in address:
#             names_checked.remove(name)
#     for name in names_checked:
#         if name in first_names:
#             names_checked.remove(name)

    return extracted_names

def replace_html(cell):
    cell.replace('<br>', ' ')
    cell.replace('\n', ' ')
    return cell


def easy_name(cell):
    """ Clear Hou at end, moms, mom, house """
    try:
        if cell.endswith('hou'):
            cell = cell.rstrip('hou')

        if cell.endswith('Hou'):
            cell = cell.rstrip('Hou')

        if cell.endswith('moms'):
            cell = cell.rstrip('moms')

        if cell.endswith('Moms'):
            cell = cell.rstrip('Moms')

        if cell.endswith('house'):
            cell = cell.rstrip('house')
        if cell.endswith('Hom'):
            cell = cell.rstrip('Hom')
        if type(cell) == 'NoneType':
            cell = ''
    except:
        return cell
    
    return cell

def last_name(cell):
    try:
        cell = cell.split()
        if len(cell) > 1:
            last_name = cell[-1]
            return last_name
        else:
            return ''.join(cell)
    except:
        return cell
    
def first_name(cell):
    try:
        full = cell.split()
        first = ' '.join(full[: len(full) - 1])
        return first
    except:
        return cell

def cap_words(s):
    s = string.capwords(s)
    return s
```
